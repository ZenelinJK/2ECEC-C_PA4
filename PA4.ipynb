# LuzViMinda Board Exam

```python
import pandas as pd # Import Pandas Library

# Read xlsx File Using .read_excel Into a Dataframe for Instru
Instru = pd.read_excel('board2.xlsx')

# Filter Constant Attributes: Instrumintation and Luzon. Electronics > 70 is also a Constant that can be Read by the System
LuzInstru = Instru[(Instru['Electronics']>70)&(Instru['Track'] == 'Instrumentation')&(Instru['Hometown'] == 'Luzon')]

# Store Results into Luzo with Expected Output for Name, GEAS and Electronics alone
Luzo = LuzInstru[['Name', 'GEAS', 'Electronics']]

#Divider üëç
divider = '=' * 50

# Read xlsx File using .read_excel Into a Dataframe for Mindy
Mindy = pd.read_excel('board2.xlsx')

# Get Average of All Grades of All Subjects
Mindy['Average'] = Mindy[['Math','Electronics','GEAS','Communication']].mean(axis=1) # Axis=1 to load the values into Average

# Filter Constant Attributes: Electronics, Mindanao and Female
MindMindy = Mindy[(Mindy['Electronics'])&(Mindy['Hometown'] == 'Mindanao')&(Mindy['Gender'] == 'Female')]

# Store Results into Midna with Expected Output for Name, Tracks, Electronics and Average alone
Minda = MindMindy[['Name', 'Track', 'Electronics', 'Average']]

# Display Results for Instru and Mindy
Luzo, divider, Minda
```

# Grade Visualization

```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read and Store Dataframe
Grades = pd.read_excel('board2.xlsx')

# Calculate Average Grade
Grades['Average_Grade'] = Grades[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

# Average Grade per Track via Barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Track', y='Average_Grade', data=Grades)
plt.title('Average Grade by Track')
plt.show()

# Box Plot Average Grade per Track
plt.figure(figsize=(12, 6))
sns.boxplot(x='Track', y='Average_Grade', data=Grades)
plt.title('Box Plot of Average Grade by Track')
plt.show()

# Average Grade per Gender via Barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Gender', y='Average_Grade', data=Grades)
plt.title('Average Grade by Gender')
plt.show()

# Box Plot Average Grade per Gender
plt.figure(figsize=(12, 6))
sns.boxplot(x='Gender', y='Average_Grade', data=Grades)
plt.title('Box Plot of Average Grade by Gender')
plt.show()

# Average Grade per Hometown via Barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Hometown', y='Average_Grade', data=Grades)
plt.title('Average Grade by Hometown')
plt.show()

# Box Plot Average Grade per Hometown
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hometown', y='Average_Grade', data=Grades)
plt.title('Box Plot of Average Grade by Hometown')
plt.show()
```
